## Detecting Outlier using Percentile & Remove them

```python
max_threshold = df['price'].quantile(0.99)
```

```python
max_threshold
```

```python
min_threshold = df['price'].quantile(0.05)
```

```python
df_new = df[(df['price']>min_threshold) & (df['price']<max_threshold)]
```

```python
df_new.sample(10)
```
