```python
import pandas as pd
```

```python
df = pd.read_csv('Book1.csv')
df
```

```python
df = df.drop(['Unnamed: 0'],axis=1)
df
```

```python
dict = ({
    'temperature':df.temperature.mean(),
    'windspeed':df.windspeed.mean()
})
ndf = df.fillna(value=dict)
ndf
```

```python
dict = ({
    'temperature':-9999,
    'windspeed':-9999
})
ndf1 = df.fillna(value=dict)
ndf1
```

```python
ndf1
```

```python
ndf2 = ndf1.set_index('day')
```

```python
ndf2
```
